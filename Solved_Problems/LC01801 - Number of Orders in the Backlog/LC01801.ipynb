```python
import heapq
from typing import *

class Solution:
    def getNumberOfBacklogOrders(self, orders: List[List[int]]) -> int:
        
        buy_backlog = []
        sell_backlog = []
        
        for price, amount, orderType in orders:
            
            tmp = []
            if orderType == 0: # buy
                
                while sell_backlog:
                    sell_price, sell_amount = heapq.heappop(sell_backlog)
                    
                    if sell_price <= price:
                        if sell_amount > amount:
                            heapq.heappush(sell_backlog, (sell_price, sell_amount - amount))
                            amount = 0
                            break
                        else:
                            amount -= sell_amount
                    else:
                        heapq.heappush(sell_backlog, (sell_price, sell_amount))
                        break
                        
                if amount:
                    heapq.heappush(buy_backlog, (-price, amount))
                    
            else: # sell
                
                while buy_backlog:
                    buy_price, buy_amount = heapq.heappop(buy_backlog)
                    buy_price *= -1
                    
                    if buy_price >= price:
                        if buy_amount > amount:
                            heapq.heappush(buy_backlog, (-buy_price, buy_amount - amount))
                            amount = 0
                            break
                        else:
                            amount -= buy_amount
                    else:
                        heapq.heappush(buy_backlog, (-buy_price, buy_amount))
                        break
                    
                if amount:
                    heapq.heappush(sell_backlog, (price, amount))
        
        ans = 0
        for _, amount in buy_backlog:
            ans += amount
            ans %= 10**9 + 7
        for _, amount in sell_backlog:
            ans += amount
            ans %= 10**9 + 7
        return ans
```
